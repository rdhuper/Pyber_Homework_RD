```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_to_load= pd.read_csv("data/city_data.csv")
ride_data_to_load= pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
combined_data= pd.merge(ride_data_to_load, city_data_to_load,how= 'outer', on= "city")


# Display the data table for preview
combined_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# seperate dataframes by city type
urban = combined_data.loc[(combined_data["type"] == "Urban")]
suburban = combined_data.loc[(combined_data["type"] == "Suburban")]
rural = combined_data.loc[(combined_data["type"] == "Rural")]



```

```python
#calculations for urban 
urban_grouped = urban.groupby(['city'])
avg_fare_urban= urban_grouped['fare'].mean()
total_rides_urban=urban_grouped['ride_id'].count()
total_urban_drivers= urban_grouped['driver_count'].mean()

#calculations for suburban
suburban_grouped = suburban.groupby(['city'])
avg_fare_suburban= suburban_grouped['fare'].mean()
total_rides_suburban=suburban_grouped['ride_id'].count()
total_suburban_drivers= suburban_grouped['driver_count'].mean()

#calculations for rural
rural_grouped = rural.groupby(['city'])
avg_fare_rural= rural_grouped['fare'].mean()
total_rides_rural=rural_grouped['ride_id'].count()
total_rural_drivers= rural_grouped['driver_count'].mean()


```

```python
# Build the scatter plots for each city types
ax = plt.subplots()


#urban
plt.scatter(total_rides_urban, avg_fare_urban, marker="o", facecolors="lightcoral", edgecolors="black", 
            s = 5*total_urban_drivers,label="Urban", alpha= 0.5, linewidth=1)

#suburban
plt.scatter(total_rides_suburban, avg_fare_suburban, marker="o", facecolors="lightskyblue", edgecolors="black", 
            s = 5*total_suburban_drivers, label="Suburban", alpha= 0.5, linewidth=1)

#rural
plt.scatter(total_rides_rural, avg_fare_rural, marker="o", facecolors="gold", edgecolors="black", 
            s = 5* total_urban_drivers, label = "Rural", alpha=0.5, linewidth=1)

plt.grid(True)

plt.title("Pyber Ride Sharing Data (2016)")

plt.xlabel("Total Number of Rides (Per City)")

plt.ylabel("Average Fare ($)")

plt.legend(loc='upper right')

plt.suptitle("Note:" + "\n" + "Circle size correlates with driver count per city ", x=1.3, y=0.5)

plt.savefig("Images/Pyber Ride Sharing Data (2016).png")

plt.show()



```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fare= combined_data.groupby(["type"])["fare"].sum()

# Build Pie Chart
labels= ["Rural", "Suburban", "Urban"]

colors= ["gold", "lightskyblue", "lightcoral"]

explode= [0, 0, 0.1]

plt.pie(total_fare, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=160)
plt.title(" % of Total Fares by City")


plt.savefig("Images/% of Total Fares by City.png")

plt.show()


```

```python
# Calculate Ride Percents
total_rides= combined_data.groupby(["type"])["ride_id"].count()

# Build Pie Chart
labels= ["Rural", "Suburban", "Urban"]

colors= ["gold", "lightskyblue", "lightcoral"]

explode= [0, 0, 0.1]

plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=160)
plt.title(" % of Rides by City Type")

plt.savefig("Images/% of Rides By City Type.png")

plt.show()





```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_driver_percent= combined_data.groupby(["type"])["driver_count"].sum()

# Build Pie Chart
labels= ["Rural", "Suburban", "Urban"]

colors= ["gold", "lightskyblue", "lightcoral"]

explode= [0, 0, 0.1]

plt.pie(total_driver_percent, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=160)
plt.title(" % of Total Drivers by City Type")

plt.savefig("Images/% of Total Drivers By City Type.png")

plt.show()



```

# Conclusion about the dataset:

1. First conclusion that can be made by looking at the Scatter Plot is that although rural cities have much less rides per city the average fare is much higher than that of urban cities. 
2. Suburban cities have the lowest driver count out of all 3 categories which could lead drivers to earn more as there is not as much compeition. 
3. In Contrast, the number of drivers in Urban cities is far more than suburban and rural which leads to higher competition and lower rides per driver.
